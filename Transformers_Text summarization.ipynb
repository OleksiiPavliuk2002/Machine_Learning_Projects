```python
from transformers import pipeline
```

```python
# Загрузка предварительно обученной модели для генерации текста
model_name = "t5-base"  # Измените на нужную модель, например, "t5-small"
summarizer = pipeline("summarization", model=model_name, tokenizer=model_name)
```

```python
# Входной текст, для которого нужно сгенерировать аннотацию
input_text ='As you know we have four seasons in our country: winter, spring, summer and autumn. All of them are great in their own way. In winter we have a lot of snow, holidays and gifts. Spring is full of new life and warmth. I like autumn because of multi-colored leaves and rainy weather. But my favorite season is summer'
```

```python
# Генерация аннотации с помощью предварительно обученной модели
summary = summarizer(input_text, max_length=100, min_length=30, do_sample=False)
```

```python
# Вывод результата
print("Сгенерированная аннотация:")
print(summary[0]['summary_text'])
```
